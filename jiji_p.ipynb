```python
from bs4 import BeautifulSoup as soup
import urllib
from urllib.request import urlopen, Request
my_url="https://jiji.ng/mobile-phones-tablets"
header = {
      'User-Agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Mobile Safari/537.36'}
req = urllib.request.Request(url=my_url, headers=header) 
page_html = urllib.request.urlopen(req).read()
if page_html:
    print('Connected')
else:
    print('Not connected')
```

```python
page_soup = soup(page_html, "html.parser")
#page_soup
```

```python
containers = page_soup.findAll("div", { "class": "masonry-wall b-list-advert__gallery"})
len(containers)
#print(soup.prettify(containers[0]))
```

```python
container = containers[0]
contents = container.findAll("div", {"class": "masonry-item"})
len(contents)
```

```python
prod_contents = contents[0]
product_image = prod_contents.findAll("source", {"type": "image/webp"})
print(product_image[0]['srcset'])
```

```python
product_name = prod_contents.findAll("div", {"class": "b-advert-title-inner qa-advert-title b-advert-title-inner--div"})
print(product_name[0].text)
```

```python
product_price= prod_contents.findAll("div", {"class": "qa-advert-price"})
print(product_price[0].text)
```

```python
product_condition= prod_contents.findAll("div", {"class": "b-list-advert-base__item-attr"})
print(product_condition[0].text)
```

```python
product_verification= prod_contents.findAll("div", {"class": "b-list-advert-base__label__inner"})
print(product_verification[0].text)
```

```python
import pandas as pd 
import numpy as np
```

```python
product_image = []
product_name = []
product_price=[]
product_condition= []
product_verification= []


for content in contents:
    image = content.findAll("source", {"type": "image/webp"})
    if len(image) == 0:
        product_image.append(np.nan)
    else:
        product_image.append(image[0]['srcset'])
    name =content.findAll("div", {"class": "b-advert-title-inner qa-advert-title b-advert-title-inner--div"})
    if len(name) == 0:
        product_name.append(np.nan)
    else:
        product_name.append(name[0].text)
    price = content.findAll("div", {"class": "qa-advert-price"})
    if len(price) == 0:
         product_price.append(np.nan)
    else:
         product_price.append(price[0].text)
    condition = content.findAll("div", {"class": "b-list-advert-base__item-attr"})
    if len(condition) == 0:
        product_condition.append(np.nan)
    else:
         product_condition.append(condition[0].text)
    verification = content.findAll("div", {"class": "b-list-advert-base__label__inner"})
    if len(verification) == 0:
         product_verification.append(np.nan)
    else:
         product_verification.append(verification[0].text)



print(len(product_image ))
print(len(product_name ))
print(len(product_price))
print(product_condition)
print(product_verification)
```

```python
df = pd.DataFrame({'Image_Link':product_image,'Product_name':product_name,'Price':product_price,"Product_condition":product_condition,
                   'Product_verification':product_verification})
df.to_csv('Jiji.csv')
Jiji = pd.read_csv('Jiji.csv')
Jiji
```
